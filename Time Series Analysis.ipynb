```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sample_Superstore.xls - Orders.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df['Category'].value_counts()
```

```python
furniture = df[df["Category"]=="Furniture"]
```

```python
furniture.shape
```

```python
furniture.dtypes
```

```python
furniture['Order Date'] = pd.to_datetime(furniture['Order Date'])
```

```python
furniture['Order Date'].min()
```

```python
furniture['Order Date'].max()
```

```python
furniture = furniture[['Order Date','Sales']]
```

```python
furniture.shape
```

```python
furniture = furniture.sort_values('Order Date',ascending=True)
```

```python
furniture.isnull().sum()
```

```python
furniture.head()
```

```python
furniture['Order Date'].value_counts()
```

```python
furniture = furniture.groupby('Order Date')['Sales'].sum().reset_index()
```

```python
furniture.shape
```

```python
furniture.head()
```

```python
furniture = furniture.set_index('Order Date')
```

```python
furniture
```

```python
y = furniture['Sales'].resample("M").mean()
```

```python
y.shape
```

```python
y.plot(figsize=(10,6))
plt.show()
```

```python
train = y.loc[:'2016-12-31']
print(train.shape)
train
```

```python
test = y.loc['2017-01-31':]
print(test.shape)
test
```

```python
train.plot(figsize=(10,6),title='Average Sales',fontsize=14)
test.plot(figsize=(10,6),title='Average Sales',fontsize=14)
plt.show()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
Exp_Smooth = test.copy()
```

```python
fit1 = SimpleExpSmoothing(train).fit(smoothing_level = 0.01)
Exp_Smooth['SES'] = fit1.forecast(steps=len(test))
```

```python
train.plot(figsize=(10,6),title='Average Sales',fontsize=14)
test.plot(figsize=(10,6),title='Average Sales',fontsize=14)
Exp_Smooth['SES'].plot(figsize=(10,6),title='Average Sales',fontsize=14)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt
```

```python
rmse = sqrt(mean_squared_error(test,Exp_Smooth.SES))
print(rmse)
```

# HOLT MODEL

```python
#Holt Model
from statsmodels.tsa.api import Holt

holt_df = test.copy()
```

```python
fit2 = Holt(train).fit(smoothing_level = 0.01,smoothing_slope=0.25)
holt_df['Holt_linear'] = fit2.forecast(steps=len(test))
```

```python
train.plot(figsize=(10,6),title='Average Sales',fontsize=14)
test.plot(figsize=(10,6),title='Average Sales',fontsize=14)
holt_df['Holt_linear'].plot(figsize=(10,6),title='Average Sales',fontsize=14)
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test,holt_df['Holt_linear']))
print(rmse)
```

# HOLT WINTER MODEL

```python
from statsmodels.tsa.api import ExponentialSmoothing
holt_winter_df = test.copy()
```

```python
fit3 = ExponentialSmoothing(train , seasonal_periods=12,
                          trend='add',
                          seasonal='add').fit()
holt_winter_df['holt_winter'] = fit3.forecast(steps=len(test))
```

```python
train.plot(figsize=(10,6),title='Average Sales',fontsize=14)
test.plot(figsize=(10,6),title='Average Sales',fontsize=14)
holt_winter_df['holt_winter'].plot(figsize=(10,6),title='Average Sales',fontsize=14)
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test,holt_df['Holt_linear']))
print(rmse)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
fit1 = ExponentialSmoothing(y, seasonal_periods=12, trend='add',seasonal='add').fit()
```

```python
y_pred = fit1.forecast(steps=12)
y.plot(figsize=(10,6), title='Average Sales', fontsize=14)
y_pred.plot(figsize=(10,6), title='Average Sales',fontsize=14)
plt.show()
```

```python
rmse = sqrt(mean_squared_error(test,y_pred))
print(rmse)
```
